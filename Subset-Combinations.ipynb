```python
'''
We are doing binary selection (0/1) here. For each num, I can select or not select
We are looking for all the leaf node on the search tree.

The defintion of recursion
nums: nums
index: which number we are operating on.
path: the current selection we have upto this index.
        [] when index == 0
results: results
'''
class Solution:
    """
    @param nums: A set of numbers
    @return: A list of lists
    """
    def subsets(self, nums):
        nums.sort()
        results = []
        self.dfs(nums, 0, [], results)
        return results
        
    
    def dfs(self, nums, index, path, results):
        '''When we get to the last num, we have made a selection'''
        if index == len(nums):
            results.append(path[:])
            return
        
        '''Select the current number'''
        path.append(nums[index])
        self.dfs(nums, index + 1, path, results)
        
        '''Not select the current number'''
        path.pop()  # Remove the selected number
        self.dfs(nums, index + 1, path, results)
        
        
'''
The search tree here has all the nodes as answers
DFS definition:
    nums: nums
    index: starting index for the searching
    path: the current selection at the current index
'''                
class Solution:
    """
    @param nums: A set of numbers
    @return: A list of lists
    """
    def subsets(self, nums):
        nums.sort()
        results = []
        self.dfs(nums, 0, [], results)
        return results
        
    
    def dfs(self, nums, index, path, results):
        # Every node is an answer
        results.append(path[:])
        
        for i in range(index, len(nums)):
            path.append(nums[i])
            self.dfs(nums, i + 1, path, results)
            path.pop()
        
```
