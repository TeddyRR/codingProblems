```python
"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None
"""

class Solution:
    """
    @param root: the root of binary tree
    @return: the length of the longest consecutive sequence path
    """
    def longestConsecutive(self, root):
        return self.helper(root, root.val, 0)
        
    
    '''p_val is the parent value'''
    def helper(self, root, p_val, cur_L):
        if not root:
            return 0
        
        if root.val == p_val + 1:
            cur_L += 1
        else:
            cur_L = 1
        
        return max(self.helper(root.left, root.val, cur_L), 
                    self.helper(root.right, root.val, cur_L), 
                    cur_L)
        
```

```python
"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None
"""

class Solution:
    """
    @param root: the root of binary tree
    @return: the length of the longest consecutive sequence path
    """
    def longestConsecutive2(self, root):
        '''
            Definition of the three return values
            1. The max length of the sequence in the tree
            2. The max length of the increasing sequence starting from the root
            3. The max length of the decreasing sequence starting from the root
        '''
        max_len, _, _, = self.helper(root)
        
        return max_len
    
    
    def helper(self, root):
        if not root:
            return 0, 0, 0
        
        left_max_len, left_dec_from_root, left_inc_from_root = self.helper(root.left)
        right_max_len, right_dec_from_root, right_inc_from_root = self.helper(root.right)
        
        dec_from_root, inc_from_root = 0, 0
        
        if root.left:
            if root.left.val + 1 == root.val:
                dec_from_root = left_dec_from_root + 1
                
            if root.left.val - 1 == root.val:
                inc_from_root = left_inc_from_root + 1
        
        if root.right:
            if root.right.val + 1 == root.val:
                dec_from_root = max(dec_from_root, right_dec_from_root + 1)
                
            if root.right.val - 1 == root.val:
                inc_from_root = max(inc_from_root, right_inc_from_root + 1)
        
        len_thru_root = dec_from_root + inc_from_root + 1
        
        max_len = max(len_thru_root, left_max_len, right_max_len)
        
        return max_len, dec_from_root, inc_from_root
        
```

```python
"""
Definition for a multi tree node.
class MultiTreeNode(object):
    def __init__(self, x):
        self.val = x
        children = [] # children is a list of MultiTreeNode
"""


class Solution:
    # @param {MultiTreeNode} root the root of k-ary tree
    # @return {int} the length of the longest consecutive sequence path
    def longestConsecutive3(self, root):
        
        '''
            Definition of the three return values
            1. The max length of the sequence in the tree
            2. The max length of the increasing sequence starting from the root
            3. The max length of the decreasing sequence starting from the root
        '''
        max_len, _, _, = self.helper(root)
        
        return max_len
    
    
    def helper(self, root):
        if not root:
            return 0, 0, 0
            
        max_len, dec_from_root, inc_from_root = 0, 0, 0
        
        '''Similar to the binary tree, here we also do a pre-order traversal'''
        for child in root.children:
            result = self.helper(child)
            
            max_len = max(max_len, result[0])
            
            if child.val + 1 == root.val:
                dec_from_root = max(dec_from_root, result[1] + 1)
            
            if child.val - 1 == root.val:
                inc_from_root = max(inc_from_root, result[2] + 1)
            
        
        len_thru_root = dec_from_root + inc_from_root + 1
        
        max_len = max(len_thru_root, max_len)
        
        return max_len, dec_from_root, inc_from_root
```
